```python
#1.
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder

# Encode categorical variables
label_encoders = {}
for column in df.columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le

# Split the dataset
X = df.drop('Play', axis=1)
y = df['Play']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the model
model = GaussianNB()
model.fit(X_train, y_train)

# Make predictions
predictions = model.predict(X_test)

# Display predictions
print("Predictions:", predictions)

#2.
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split

# Vectorize the email text
vectorizer = CountVectorizer()
X_spam = vectorizer.fit_transform(df_spam['Email'])

# Split the dataset
y_spam = df_spam['Label']
X_train_spam, X_test_spam, y_train_spam, y_test_spam = train_test_split(X_spam, y_spam, test_size=0.3, random_state=42)

# Train the model
spam_model = MultinomialNB()
spam_model.fit(X_train_spam, y_train_spam)

# Make predictions
spam_predictions = spam_model.predict(X_test_spam)

# Display predictions
print("Spam Predictions:", spam_predictions)



import numpy as np

# AND gate truth table
X_and = np.array([[0, 0],
                  [0, 1],
                  [1, 0],
                  [1, 1]])
y_and = np.array([0, 0, 0, 1])  # Output for AND gate

# OR gate truth table
X_or = np.array([[0, 0],
                 [0, 1],
                 [1, 0],
                 [1, 1]])
y_or = np.array([0, 1, 1, 1])  # Output for OR gate

class Perceptron:
    def __init__(self, learning_rate=0.1, epochs=100):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.epochs):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = self.activation_function(linear_output)

                # Update weights and bias
                update = self.learning_rate * (y[idx] - y_predicted)
                self.weights += update * x_i
                self.bias += update

    def activation_function(self, x):
        return np.where(x >= 0, 1, 0)

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        return self.activation_function(linear_output)

# Train the Perceptron for AND gate
perceptron_and = Perceptron(learning_rate=0.1, epochs=10)
perceptron_and.fit(X_and, y_and)

# Train the Perceptron for OR gate
perceptron_or = Perceptron(learning_rate=0.1, epochs=10)
perceptron_or.fit(X_or, y_or)

# Test the AND gate
print("AND Gate Predictions:")
for x in X_and:
    print(f"Input: {x}, Predicted: {perceptron_and.predict(x)}")

# Test the OR gate
print("OR Gate Predictions:")
for x in X_or:
    print(f"Input: {x}, Predicted: {perceptron_or.predict(x)}")




```
